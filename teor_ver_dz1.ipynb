```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

*1) Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.*


```python
print ("a)")
m=combinations(13,4)
n = combinations(52,4)
print("все комбинации выбора крестовой карты при извлечении 4 карт: ",m)
print("все комбинации выбора 4 карт из колоды: ", n)
print("вероятность того, что все карты крести: ", round(m/n*100,1), "процента")
```

```python
print ("b)")
m=combinations(4,1)
n = combinations(49,4)
o = combinations(52,5)
print("все комбинации выбора туза из 4 тузов: ",m)
print("все комбинации выбора 4 карт из колоды без тузов: ", n)
print("все комбинации туза и остальных карт: ",m* n)
print("все комбинации выбора 5 карт из колоды: ",o)
print("вероятность, что среди 4-х карт окажется хотя бы один туз: ", round(m*n/o*100,1), "процента")
```

*4) В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?*

```python
m=combinations(2,2)
n = combinations(100,2)
print("все комбинации выбрать 2 билета: ",m)
print("все комбинации 2-х выигрышных билетов из 100: ", n)
print("вероятность того, что эти 2 билета выиграют: ", round(m/n*100,3), "процента")
```
