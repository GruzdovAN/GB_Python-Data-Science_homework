```python
from math import factorial
import numpy as np
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

*1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.*


$$P_n(X=k) = C_n^k p^k q^{n-k}$$

```python
n = round(combinations(100,85)*(0.8**85)*0.2**(100-85)*100,1)
print("Данная вероятность является дискретной, поэтому расчет вероятности попадания нужно произвести по формуле Бернулли.")
print("Итого вероятность того, что стрелок попадет в цель ровно 85 раз равна", n, 'процента')
```

*2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?*

```python
n = 5000
m = 0
p = 0.0004
lambda_ = n * p

p_0  = round((lambda_**m / np.math.factorial(m)) * np.exp(-lambda_)*100,1)
```

```python
n = 5000
m = 2
p = 0.0004
lambda_ = n * p

p_2  = round((lambda_**m / np.math.factorial(m)) * np.exp(-lambda_)*100,1)
```

```python
print("Данная вероятность является неприрывной, поэтому расчет вероятности попадания нужно произвести по формуле Пуассона.")
print("Итого вероятность того, ни одной лампочки не перегорит в 1 день составляет", p_0, 'процента')
print("Итого вероятность того, 2 лампочки перегорит в 1 день составляет", p_2, 'процента')
```

*3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?*

```python
n = round(combinations(144,70)*(0.5**70)*0.5**(144-70)*100,1)
print("Данная вероятность является дискретной, поэтому расчет вероятности попадания нужно произвести по формуле Бернулли.")
print("из опытов известно, что вероятность выпаданиея орла равноа 0,5")
print("Итого вероятность того, что орел выпадет ровно 70 раз", n, 'процента')
```

*4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?*

```python
b1= 1/2*7/10+1/2*9/11 # выбор первого мяча из 2-х корзин
b2= 1/2*6/9+1/2*8/10 #выбор второго мяча из 2-х корзин
print("По формуле произведения события вероятность того, что все мячи белые равна: ", round(b1*b2*100,1),'процентов')
print("По формуле суммы событий вероятность того, что хотя бы один мяч белый равна: ", round((b1+b2-b1*b2)*100,1),'процентов')
```
